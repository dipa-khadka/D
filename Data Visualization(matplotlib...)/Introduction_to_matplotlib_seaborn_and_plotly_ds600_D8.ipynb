```python
# golo (points) lai sano thulo banaune
plt.figure(figsize=(4,2)) # controls the size of a figure to be plotted
plt.plot(months,growth,'o-', color='purple' , ms=10) # ms vaneko marker size o ko size ho
plt.xlabel("Months")
plt.ylabel("Growth (%)")
plt.title("Monthly Growth in (%)")
plt.show()
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun']
actual_growth = [3,3.5,5,5.9,7,9]
predicted_growth = [2,2.7,4,5.5,7.6,8.9]
```

```python
# yeutai graph each actual_growth and predicted_growth plot garna
plt.plot(months, actual_growth, 'o-')
plt.plot(months, predicted_growth,'o-')
```

```python
# to know which color is actual or which one color is predected
plt.plot(months, actual_growth, 'o-', label='Actual')
plt.plot(months, predicted_growth,'o-', label='Predicted')
plt.xlabel("Months")
plt.ylabel("Growth (%)")
plt.legend() # plt.legend le plt.plot() ko table plot garxa
plt.show()
```

# **Barplot**

```python
gender = ["Male", "Female"]
frequency = [157,87]
```

```python
plt.bar(x=gender, height=frequency)
```

```python
plt.bar(x=gender, height=frequency)
plt.xlabel("Gender")
plt.ylabel("Frequency/Count")
plt.show()
```

```python
plt.bar(x=gender, height=frequency , color=['orange','red'])
plt.xlabel("Gender")
plt.ylabel("Frequency/Count")
plt.show()
```

```python
plt.bar(x=gender, height=frequency , color=['tab:orange', 'tab:red']) # tab le chahi gadha color dinxa
plt.xlabel("Gender")
plt.ylabel("Frequency/Count")
plt.show()
```

```python
plt.bar(x=gender, height=frequency , color=['tab:blue', 'tab:pink']) # tab le chahi gadha color dinxa
plt.xlabel("Gender")
plt.ylabel("Frequency/Count")
plt.show()
```

# **Barplot using seaborn**

```python
import seaborn as sns # seaborn and matplotlib are same
```

```python
sns.barplot(x=gender, y=frequency) # matplotlib ma height hunxa vane seaborn ma chahi y lekhnu parxa
```

```python
sns.barplot(x=gender, y=frequency)
plt.xlabel('Gender')
plt.ylabel("Frequency/Count")
plt.show()
```

# **Barplot using plotly**

```python
import plotly.express as px
```

```python
fig = px.bar(x=gender, y=frequency)
fig.update_layout(height=300, width=400)
```

```python
fig = px.bar(x=gender, y=frequency , text=frequency) # text=frequency le bar ko top ma value dekhauxa
fig.update_layout(height=300, width=400)
```

```python
fig = px.bar(x=gender, y=frequency , text=frequency)
fig.update_layout(height=300, width=400, xaxis_title="Gender", yaxis_title="Freuency/Count")
```

# **Pie chart**

```python
# plotly bata pie chart ramro aauxa
gender = ["Male", "Female"] # using matplotlib
frequency=[157,87]
```

```python
plt.pie(x=frequency, labels=gender)
plt.show()
```

```python
plt.pie(x=frequency, labels=gender ,autopct= '%.3f%%') # point paxadi kat ota foating rakhne
plt.show()
```

```python
px.pie(names=gender, values=frequency) # pie chart using plotly
```
