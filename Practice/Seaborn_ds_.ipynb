```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the data set
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
# lets make line plot on signal column and timepoint column to know how the value of signal is vary with timepoint value :
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.title('line plot between timplot and signal columns')
plt.show()
```

```python
# frouping this on the basis of even
sns.lineplot(x='timepoint',y='signal',hue='event',data=fmri)
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',hue='event',data=fmri,style='event',markers='o',size=100)
plt.grid(True)
plt.show()
```

## Seaborn Barplot :

```python
sns.set(style='whitegrid')
pokemon=pd.read_csv('/content/pokemon.csv')
pokemon.head()
```

```python
sns.barplot(x='is_legendary',y='speed',data=pokemon)
plt.show()
```

```python
sns.barplot(x='is_legendary',y='speed', hue='generation',data=pokemon)
plt.show()
```
