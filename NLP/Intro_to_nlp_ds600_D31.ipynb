# **Steaming :-**
It converts words into root word or base word stem or word-word may not have any meaning-very fast.

```python
words = ['change','changes','changed']
```

```python
# finding out root word of words
```

```python
stemmer = nltk.stem.PorterStemmer()
```

```python
[stemmer.stem(word)for word in words ]
```

```python
# stop words applying
d1 = "He is a lazy boy. She is also lazy."
d2 = "Neeraj is a lazy person"

corpus = [d1,d2]

vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(corpus)
vocab = vectorizer.get_feature_names_out()
pd.DataFrame(X.toarray(),columns=vocab)
```

# **Tfidf:-**


```python
# countvectorization
d1 = "he is a good boy "
d2 = "she is a good girl"
d3 = "boy and girl are good"
d4 = "good boy and good girl are good for the society"

corpus = [d1,d2,d3,d4]


vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(corpus)
vocab = vectorizer.get_feature_names_out()
pd.DataFrame(X.toarray(),columns=vocab)
```

Count vectorization gives more weight to frequently occuring word.

for eg. the keywords might appear only one/few time so we might want to boost the importance of rare occuring word. In such case countvectorizer fails. Therefore we use Tfidf vectorization which gives weight to rare occuring words.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Tfidf
d1 = "he is a good boy "
d2 = "she is a good girl"
d3 = "boy and girl are good"
d4 = "good boy and good girl are good for the society"

corpus = [d1,d2,d3,d4]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
vocab = vectorizer.get_feature_names_out()
pd.DataFrame(X.toarray(),columns=vocab)
```

```python
# stopwords
# Tfidf
d1 = "he is a good boy "
d2 = "she is a good girl"
d3 = "boy and girl are good"
d4 = "good boy and good girl are good for the society"

corpus = [d1,d2,d3,d4]

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(corpus)
vocab = vectorizer.get_feature_names_out()
pd.DataFrame(X.toarray(),columns=vocab)
```
