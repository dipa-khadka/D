```python
import pandas as pd
```

```python
# making dictionary
data = {
    'youtubers':['KSI', 'Mr.Beast', 'Sisan Baniya', 'MRB','Savya', 'Surakshya KC'],
    'gender':['Male','Male','Male','Male','Female','Female']
}
```

```python
df = pd.DataFrame(data)
df
```

```python
def clean_gender(gen):
  return gen[0]
```

```python
df # apply matra garera df unpdate hudaina
```

```python
df
```

```python
# Now i want to drop gender column
df.drop(columns=['gender']) # yesle original dataframe df update gardaina
```

```python
df
```

```python
# aaba drop garda column auto drop hunu paryo dataframe bata, yesko lagi inplace = True lekhne
df.drop(columns=['gender'], inplace= True) #apply ma inplace hudaina /inplace = True le dataframe update garxa
```

```python
df
```

```python
#  I want to rename clean gender to gender & youtubers to youtube channel
df.rename({'clean_gender':'gender', 'youtubers':'youtube channel'})
```

```python
df.rename({'clean_gender':'gender', 'youtubers':'youtube channel'}, axis = 'columns')
```
