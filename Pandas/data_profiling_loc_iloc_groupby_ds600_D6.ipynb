# loc vs iloc

```python
df.head(2)
```

```python
df.loc[12:14,['total_bill','tip']]
```

```python
# data slice garne and condition lagaune kam loc le garxa
# 14 row pani aayo ? loc vaneko location ho index hoina; loc le last ko value pani linxa
```

```python
# loc rule:
# 1. loc ma column ko thau ma index dina paudaina column name nai dinu parxa in lis
df.loc[12:14,0:2] # 0:2 lekhna paudaina loc expects column name instead not a index
```

```python
# column ma name navayera index lekhnu paryo vane iloc use garne
```

```python
df.iloc[12:14, 0:2]
```

```python
# iloc le row ra column ma last ko value lidaina, iloc expects index in both row and column
```

```python
# last 2  column and first five rows
df.iloc[0:5,-2:]
```
