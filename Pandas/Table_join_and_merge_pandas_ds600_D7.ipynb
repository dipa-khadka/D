```python
import pandas as pd
```

```python
# execel sheet read garne
emp = pd.read_excel('/pandas/employee-dept-table.xlsx', sheet_name='employee')
emp
```

```python
department = pd.read_excel('/content/employee-dept-table.xlsx', sheet_name='dept')
department
```

```python
# Inner join
pd.merge(left=emp, right=department, on = 'DepartmentID', how='inner')
```

```python
# outer join
pd.merge(left=emp, right=department, on = 'DepartmentID', how='outer')
```

```python
# right outer join
pd.merge(left=emp, right=department, on = 'DepartmentID', how='right')
```

```python
# left outer join
pd.merge(left=emp, right=department, on = 'DepartmentID', how='left')
```
