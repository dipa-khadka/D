```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head(2)
```

```python
del df['species'] # removing the target y which is species
```

```python
df.head(2)
```

```python
# finding out species by k-means clustering
X = df.iloc[:, -2:] # features
```

```python
X.head(2)
```

```python
no_of_clusters = range(1,5)
```

```python
from sklearn.cluster import KMeans
```

```python
sse=[]
for k in no_of_clusters:
  model = KMeans(n_clusters=k, random_state=0, n_init="auto").fit(X)
  sse.append(model.inertia_)
```

```python
plt.plot(no_of_clusters, sse, 'o-',color='red')
plt.xlabel("n_clusters / K")
plt.ylabel("SSE / inertia_")
plt.show()
```

* This figure shows that k value is 3.0 by elbow method .

```python
K=3
model = KMeans(n_clusters=K, random_state=0, n_init="auto").fit(X)
clusters = model.predict(X)
clusters
```

```python
centroid = model.cluster_centers_
centroid
```

```python
sns.scatterplot(x=X.iloc[:,0],y=X.iloc[:,1], hue=clusters)
plt.show()
```

```python
sns.scatterplot(x=centroid[:,0],y=centroid[:,1],color='red' , s=100, label="Centroid", marker='v')
plt.show()
```

```python
sns.scatterplot(x=X.iloc[:,0],y=X.iloc[:,1], hue=clusters)
sns.scatterplot(x=centroid[:,0],y=centroid[:,1],color='red' , s=100, label="Centroid", marker='v')
plt.legend()
plt.show()
```
