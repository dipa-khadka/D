# **Kurtosis:**
`Kurtosis refers to the degree of presence of outliers in the distribution. Kurtosis is a statistical measure, whether the data is heavy-tailed or light-tailed in a normal distribution.

* Types of kurtosis:
1. MesoKurtic
2. PlatyKurtic
3. LeptoKurtic



> MesoKurtic:

* Normal distribution
* Person Kurtosis nearly equals to 3
* medium outliers



```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
data = np.random.normal(loc=5, scale=1, size=1500)
data
```

```python
plt.figure(figsize=(4,2))
sns.histplot(data, kde=True)
plt.show()
```

```python
from scipy.stats import kurtosis
```

```python
kurt = kurtosis(data, fisher=False) # use pearson method by disbling fisher
plt.figure(figsize=(4,2))
sns.histplot(data, kde=True)
plt.show()
```

```python
kurt = kurtosis(data, fisher=False) # use pearson method by disbling fisher
plt.figure(figsize=(4,2))
plt.title("Kurtosis: {}".format(np.round(kurt,2)))
sns.histplot(data, kde=True)
plt.show()
```



> PlatyKurtic
* Uniform distribution
* Kurtosis < 3
* low outlier



```python
np.random.seed(0)
data=np.random.uniform(low=1, high=5, size=1500)
kurt = kurtosis(data, fisher=False) # use pearson method by disbling fisher
plt.figure(figsize=(4,2))
plt.title("Kurtosis: {}".format(np.round(kurt,2)))
sns.histplot(data, kde=True)
plt.show()

```



> Leptokurtic

* laplace distribution
* kurtosis > 3
* high frequency of outlier



```python
np.random.seed(0)
data = np.random.laplace(loc=5, scale=1, size=1500)
plt.figure(figsize=(4,2))
sns.histplot(data, kde=True)
plt.show()
```
