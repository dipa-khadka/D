```python
# we want to know if there is relationship between sex and smoker :
```

```python
H0 = ' There is no relationship between sex and smoker '
H1 = ' Ther is significant relationship between sex and smoker '
```

```python
alpha = 0.05
import pandas as pd
```

```python
table = pd.crosstab (df['sex'], df['smoker'])
```

```python
table
```

```python
# This is observed table so now do chi- square test
```

```python
import numpy as np
from scipy.stats import chi2_contingency
```

```python
res = chi2_contingency(table.values)
```

```python
res
```

```python
if res.pvalue < alpha:
  print('we reject H0')
else:
  print('we fail to reject H0')
```

```python
# conclusion : there is no relationship between smoker and sex.
```

```python
expected = res.expected_freq
confidece_interval = 1-alpha
degree_of_freedom=res.dof
```

```python
from scipy.stats import chisquare
result = chisquare(table.values, expected, axis=None)
result.statistic #x square calulated
```

```python
# find chisquare tabular
import scipy
chi_sqaure_tabular = scipy.stats.chi2.ppf(confidece_interval, df=degree_of_freedom)
```
